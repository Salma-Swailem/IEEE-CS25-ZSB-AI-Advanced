# Problem 1
a) Use the numpy function [numpy.random.normal](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html) to generate a standard Gaussian random variable $X$ (zero mean and unit variance).

- Use histogram to plot the estimated PDF of the random variable $X$
- Compute the true "analytical" PDF -using the equation- of $X$ . Plot the true PDF on the top of the histogram figure.
- Calculate the sample mean and the sample variance.

### True/Analytical PDF


## using `numpy`

```python
#type your code here:
n = 1000
x = np.random.normal(0, 1, n)

sample_mean = np.mean(x)
sample_var = np.var(x)
print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_var}")

plt.hist(x, bins=50, density=True, label='Estimated PDF')

def analytical_pdf(x, mu=0, sigma=1):
    return (1 / (np.sqrt(2 * np.pi * sigma**2))) * np.exp(-(x - mu)**2 / (2 * sigma**2))

x_val = np.linspace(-4, 4, 1000)
y_val = analytical_pdf(x_val)
plt.plot(x_val, y_val, 'r-', linewidth=2, label='True PDF')
plt.grid(True)
plt.show()
```

## using `scipy`

```python
#type your code here:
sample_mean = np.mean(x)
sample_var = np.var(x)
print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_var}")

plt.hist(x, bins=50, density=True, label='Estimated PDF')

x_val = np.linspace(-4, 4, 1000)
y_val = stats.norm.pdf(x_val, loc=0, scale=1)

plt.plot(x_val, y_val, 'r-', linewidth=2, label='True PDF')

plt.title("Histogram of Standard Normal Distribution")
plt.grid(True)
plt.show()
```

- Compute the true "analytical" PDF using [scipy.stats.norm.pdf](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html). Plot the true PDF on the top of the histogram figure.
- Compute the CDF using [scipy.stats.norm.cdf](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), then plot it.

```python
#type your code here:
x_val = np.linspace(-4, 4, 1000)
pdf_val = stats.norm.pdf(x_val, loc=0, scale=1)
cdf_val = stats.norm.cdf(x_val, loc=0, scale=1)

plt.hist(x, bins=50, density=True, label='Estimated PDF')
plt.plot(x_val, pdf_val, 'r-', linewidth=2, label='True PDF')
plt.plot(x_val, cdf_val, 'g-', linewidth=2, label="True CDF")

plt.legend()
plt.grid(True)
plt.show()
```

# Problem 3: Practicing the Central Limit Theorem
Use Python to simulate the Central Limit Theorem for the following random i.i.d variables and check that the higher the number of number the variables, the more the mean of subsets behave like a gaussian random variable.


```python
population_size = 100000
sample_size = 50
num_samples = 1000 

population = np.random.uniform(low=-0.6, high=-0.4, size=population_size)

sample_means = [np.mean(np.random.choice(population, size=sample_size)) for _ in range(num_samples)]

plt.figure(figsize=(8, 5))
plt.hist(sample_means, bins=50)
plt.show()
```
