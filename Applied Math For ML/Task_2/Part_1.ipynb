# Problem 1

Use numpy to simulate the vector outcomes for repeating the following random experiments $1000$ times and use the relative frequency to compute the probability of the corresponding events:
    
1. **Experiment:** Flipping a fair coin, **Event:** A head occurs.
     
Observe how changing the number of repetition affects the computed probabilities.

**Hints:** 
- Use [numpy.random.randint](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html) or [numpy.random.choice](https://numpy.org/doc/stable/reference/random/generated/numpy.random.choice.html) to simulate the experiment.
- Other useful functions [numpy.arange](https://numpy.org/doc/stable/reference/generated/numpy.arange.html), [numpy.cumsum](https://numpy.org/doc/stable/reference/generated/numpy.cumsum.html), [numpy.sum](https://numpy.org/doc/stable/reference/generated/numpy.sum.html), and [matplotlib.pyplot.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
#type your code here:
n = [10 ,50 , 100 , 500 , 1000 , 5000, 10000, 50000, 100000]
p1 = []

for i in n:
 coin_flips = np.random.choice([0, 1], size = i)
 
 cum_heads = np.cumsum(coin_flips)
 cum_rel_freq = cum_heads / np.arange(1, i + 1)
 
 final_probability = cum_rel_freq[-1]
 p1.append(final_probability)
 
 print(f"n = {i} , P(e) = {final_probability:.5f}")


plt.plot(n, p1, marker='o', label='Simulated Probability')

plt.title('Probability of Heads vs. Num of Flips')
plt.axhline(y=0.5, color='r', linestyle='--', label='Theoretical Probability (0.5)')

plt.xscale('log')
plt.legend()
plt.grid(True)
plt.show()
```

2. **Experiment:** Rolling a fair die, **Event:** Number 5 occurs.

```python
#type your code here:
p2 = []
for i in n:
    die_rolls  = np.random.randint(1, 7, size = i)
    probability = np.sum(die_rolls == 5) / i
    p2.append(probability)
    print(f"n = {i} , P(e) = {probability:.5f}")


plt.plot(n, p2, marker='o', label='Simulated Probability')
plt.axhline(y=1/6, color='r', linestyle='--', label='Theoretical Probability (1/6)')
plt.xscale('log')
plt.title("Probability of Rolling a 5 vs. Num of Rolls")
plt.legend()
plt.grid(True)
plt.show()
```

3. **Experiment:** Rolling a fair die, **Event:** A number greater than 3 occurs.

```python
#type your code here:

p3 = []

for i in n:
    die_rolls = np.random.randint(1, 7, size=i) 
    greater_than_3 = np.sum(die_rolls > 3) 
    probability = greater_than_3 / i
    p3.append(probability)

    print(f"n = {i}, P(e) = {probability:.5f}")

plt.plot(n, p3, marker='o', label='Simulated Probability')
plt.axhline(y=0.5, color='r', linestyle='--', label='Theoretical Probability (0.5)')
plt.xscale('log')
plt.title("Probability of Rolling > 3 vs. Num of Rolls")
plt.legend()
plt.grid(True)
plt.show()
```

4. **Experiment:** Rolling two fair dice, **Event:** Sum of two dice equal 5.

```python
#type your code here:
p4 = []

for i in n:
    die1 = np.random.randint(1, 7, size = i)
    die2 = np.random.randint(1, 7, size = i)
    sum_five = np.sum((die1 + die2) == 5)
    probability = sum_five / i
    p4.append(probability)

    print(f"n = {i}, P(e) = {probability:.5f}")

plt.plot(n, p4, marker='o', label='Simulated Probability')
plt.axhline(y = 1/9 , color='r', linestyle='--', label='Theoretical Probability (1/9)')
plt.xscale('log')
plt.title("Probability of Sum = 5 vs. Number of Rolls")
plt.legend()
plt.grid(True)
plt.show()
```

# Problem 3: Independence

Use NumPy to create simulations and compute **empirical  probability** for the following outcomes. Compare the results with the **theoretical probability**.

1. Two fair coin flips produce exactly two heads

```python
#type your code here:

# simulate 1 million tests of two fair coin flips
sim_num = 10 ** 6
coin_flips = np.random.choice([0, 1], size=(sim_num, 2))

# sums of all tests
flip_sum = np.sum(coin_flips, axis=1)

# proportion of tests that produced exactly two heads
num_two_heads = np.sum(flip_sum == 2)

emp_prob = num_two_heads / sim_num
# Print results
print(f"Theoretical Probability: {1/4:.4f}")
print(f"Empirical Probability: {emp_prob:.4f}")
```

2. Three fair coin flips produce exactly one head

```python
#type your code here:

# simulate 1 million tests of three fair coin flips
coin_flips = np.random.choice([0, 1], size=(sim_num, 3))

# sums of all tests
flip_sum = np.sum(coin_flips, axis=1)

# proportion of tests that produced exactly one head
one_head = np.sum(flip_sum == 1)

emp_prob = one_head / sim_num
print(f"Theoretical Probability: {3/8:.4f}")
print(f"Empirical Probability: {emp_prob:.4f}")
```

3. Three biased coin flips with P(H) = 0.6 produce exactly one head

```python
#type your code here:


# simulate 1 million tests of three biased coin flips
# hint: use np.random.choice()
coin_flips = np.random.choice([0,1], size=(sim_num, 3), p=[0.4, 0.6])

# sums of all tests
flip_sum= np.sum(coin_flips, axis=1)

# proportion of tests that produced exactly one head
one_head = np.sum(flip_sum == 1)

empirical_prob = one_head / sim_num
theoretical_prob = 3*(0.6**1)*(0.4**2)
print(f"Theoretical Probability: {theoretical_prob:.4f}")
print(f"Empirical Probability: {empirical_prob:.4f}")
```

# Problem 5: Correlation and Covariance

The CSV file includes a hourly/daily summary for Szeged, Hungary area, between 2006 and 2016. After inspecting the data:

- Calculate the covariance and correlation coeffecient between temperature and apparent temperature
- Calculate the covariance and correlation coeffecient between temperature and humidity

```python
df = pd.read_csv('weatherHistory.csv')
temperature = df['Temperature (C)'].to_numpy().reshape(-1, 1)
apparent_temp = df['Apparent Temperature (C)'].to_numpy().reshape(-1, 1)
humidity = df['Humidity'].to_numpy().reshape(-1, 1)
```

```python
plt.scatter(temperature, apparent_temp)
plt.grid()
plt.xlabel('Temperature (C)')
plt.ylabel('Apparent Temperature (C)')
```

```python
plt.scatter(humidity, temperature)
plt.grid()
plt.xlabel('Humidity')
plt.ylabel('Temperature (C)')
```

## Temperature and Apparent Temperature

```python
#type your code here:

temp = df["Temperature (C)"]
app_temp = df["Apparent Temperature (C)"]
cov = np.cov(temp, app_temp)
corr = np.corrcoef(temp, app_temp)

print(f"Covariance: \n {cov}")
print(f"Correlation: \n {corr}")
```

## Temperature and Humidity

```python
#type your code here:
temp = df["Temperature (C)"]
humid = df["Humidity"]
cov = np.cov(temp, humid)
corr = np.corrcoef(temp, humid)

print(f"Covariance: \n {cov}")
print(f"Correlation: \n {corr}")
```
